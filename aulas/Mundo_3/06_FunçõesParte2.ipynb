# Docstrings

##### Para que a função help() retorne uma explicação sobre a função que eu criei. Usamos as docstrings. Basta inserir aspas triplas no começo e fim da string """      """

```python
def contador(i, f, p):
    """_Exibir uma contagem definindo o início, fim e valor a ser pulado_

    Args:
        i (_int_): _valor de início_
        f (_int_): _valor de fim_
        p (_int_): _valor a ser pulado_
    """
    
    c=1
    while c <= f:
        print(f'{c} ', end='..')
        c += p
    print('FIM!!')
    
    
contador(1, 100, 2)
    
```

```python
help(contador)
```

# Parâmetros opcionais

#### Podemos deixar de passar uma das variáveis para a função, caso uma seguna opção já tenha sido defenida, como exemplo em c=0. 

```python
def somar(a, b, c=0):
    s = a + b + c
    print(f'A soma vale {s}')
    
```

```python
somar(3, 2, 70)

somar(3, 2)
```
